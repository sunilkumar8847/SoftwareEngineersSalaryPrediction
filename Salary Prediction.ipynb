```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("engineers_2023.csv")
```

```python
df.head(5)
```

```python
df = df[["Country", "EdLevel", "YearsCode", "Employment", "ConvertedCompYearly", "DevType"]]  # Example adjustment
df = df.rename({"ConvertedCompYearly": "Salary"}, axis=1)
df['Salary'] = pd.to_numeric(df['Salary'], errors='coerce')

df.head()
```

```python
df = df[df["Salary"].notnull()]
df.head()
```

```python
df.info()
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
# df = df[df["Employment"] == "Employed, full-time"]
df = df[(df["Employment"] == "Employed, full-time") | (df["Employment"] == "Student, full-time") | (df["Employment"] == "Independent contractor, freelancer, or self-employed")]
df = df.drop("Employment", axis=1)
df.info()
```

```python
df["Country"].value_counts()
```

```python
def shorten_catagories(categories, cutoff):
    categorical_map = {}
    for i in range(len(categories)):
        if categories.values[i] >= cutoff:
            categorical_map[categories.index[i]] = categories.index[i]
        else:
            categorical_map[categories.index[i]] = 'Other'
    return categorical_map
```

```python
country_map = shorten_catagories(df.Country.value_counts(), 400)
df["Country"] = df["Country"].map(country_map)
df.Country.value_counts()
```

```python
df["DevType"].value_counts()
```

```python
df = df[~df["DevType"].isin(["Other (please specify):", "Student"])]
df.value_counts()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 7))
df.boxplot('Salary', 'Country', ax=ax)
plt.suptitle('Salary (US$) v Country')
plt.title('')
plt.ylabel('Salary')
plt.xticks(rotation=90)
plt.show()
```

```python
df = df[df["Salary"] <= 200000]
df = df[df["Salary"] >= 8000]
df = df[df['Country'] != 'Others']
```

```python
fig, ax = plt.subplots(1,1, figsize=(12, 7))
df.boxplot('Salary', 'Country', ax=ax)
plt.suptitle('Salary (US$) vs Country')
plt.title('')
plt.ylabel('Salary')
plt.xticks(rotation=90)
plt.show()
```

```python
df["YearsCode"].unique()
```

```python
def clean_experience(x):
    if x == 'More than 50 years':
        return 50
    elif x == 'Less than 1 year':
        return 0.5
    else:   
        return float(x)
        
df["YearsCode"] = df["YearsCode"].apply(clean_experience)

df["YearsCode"].unique()
```

```python
df ["EdLevel"].unique()
```

```python
# def clean_education(x):
#     if "Bachelorâ€™s degree" in x:
#         return "Bachelorâ€™s degree"
#     if "Masterâ€™s degree" in x:
#         return "Masterâ€™s degree"
#     if "Professional degree" in x or "Other doctoral degree" in x:
#         return "Post grad"
#     return "Less than a Bachelors"

def clean_education(x):
    if "Bachelor" in x:
        return "Bachelorâ€™s degree"
    if "Master" in x:
        return "Masterâ€™s degree"
    if "Professional" in x or "Doctoral" in x:
        return "Post grad"
    return "Less than a Bachelors"

df["EdLevel"] = df["EdLevel"].apply(clean_education)
```

```python
df["EdLevel"].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
le_education = LabelEncoder()
df['EdLevel'] = le_education.fit_transform(df["EdLevel"])
df["EdLevel"].unique()
```

```python
le_country = LabelEncoder()
df['Country'] = le_country.fit_transform(df["Country"])
df["Country"].unique()
```

```python
le_devtype = LabelEncoder()
df["DevType"] = le_devtype.fit_transform(df["DevType"])
df["DevType"].unique()
```

```python
x = df.drop("Salary", axis=1)
y = df["Salary"]
```

```python
from sklearn.linear_model import LinearRegression
linear_reg = LinearRegression()
linear_reg.fit(x, y.values)
```

```python
y_prediction = linear_reg.predict(x)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np
error = np.sqrt(mean_squared_error(y, y_prediction))
```

```python
error
```

```python
from sklearn.tree import DecisionTreeRegressor
dec_tree_reg = DecisionTreeRegressor(random_state=0)
dec_tree_reg.fit(x, y.values)
```

```python
y_pred = dec_tree_reg.predict(x)
```

```python
error = np.sqrt(mean_squared_error(y, y_pred))
print("${:,.02f}".format(error))
```

```python
from sklearn.ensemble import RandomForestRegressor
random_forest_reg = RandomForestRegressor(random_state=0)
random_forest_reg.fit(x, y.values)
```

```python
y_pred = random_forest_reg.predict(x)
```

```python
error = np.sqrt(mean_squared_error(y, y_pred))
print("${:,.02f}".format(error))
```

```python
from sklearn.model_selection import GridSearchCV

max_depth = [None, 2, 4, 6, 8, 10, 12]
parameters = {"max_depth": max_depth}

regressor = DecisionTreeRegressor(random_state=0)
gs = GridSearchCV(regressor, parameters, scoring='neg_mean_squared_error')
gs.fit(x, y.values)      
```

```python
regressor = gs.best_estimator_
regressor.fit(x, y.values)
y_pred = regressor.predict(x)
error = np.sqrt(mean_squared_error(y, y_pred))
print("${:,.2f}".format(error))
```
